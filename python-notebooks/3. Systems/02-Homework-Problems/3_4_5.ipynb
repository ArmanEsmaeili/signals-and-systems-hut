## Problem Statement

Determine whether the following system is **homogeneous** and **additive** (i.e., linear):

$$
y(t) =
\begin{cases}
x(t), & t \geq 0 \\
0,    & t < 0
\end{cases}
$$

---

## Linearity Test Approach

To determine whether the system is linear, we examine the two defining properties of linear systems: **homogeneity** and **additivity**.

Two test signals are selected and applied to the system:

- **Homogeneity:**  
  $$
  T\{\alpha x_1(t)\}\ \stackrel{?}{=}\ \alpha\, T\{x_1(t)\}
  $$
- **Additivity:**  
  $$
  T\{x_1(t)+x_2(t)\}\ \stackrel{?}{=}\ T\{x_1(t)\} + T\{x_2(t)\}
  $$

In this analysis, the signals
$$
x_1(t) = \sin(t)
\qquad
x_2(t) = \cos(t)
$$
are used over the interval $t \in [-5, 5]$.


# plot the input signal and scaled signal

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------
# Time axis
# --------------------------------------------------
t = np.linspace(-5, 5, 1000)
# Create a continuous time vector from -5 to 5

# --------------------------------------------------
# Input signals
# --------------------------------------------------
x1 = np.sin(t)
# Define first input signal x1(t) = sin(t)

x2 = np.cos(t)
# Define second input signal x2(t) = cos(t)

# --------------------------------------------------
# Scaling factor
# --------------------------------------------------
alpha = 2.0
# Define the scaling constant alpha

x1_scaled = alpha * x1
# Scale x1(t) by alpha

x2_scaled = alpha * x2
# Scale x2(t) by alpha

# --------------------------------------------------
# Create optimized scientific layout
# --------------------------------------------------
fig, axs = plt.subplots(2, 2, figsize=(12, 8), sharex=True)
# Create a 2x2 subplot layout for clear comparison

# --------------------------------------------------
# Column 1: x1(t)
# --------------------------------------------------
axs[0, 0].plot(t, x1, color='blue', linewidth=2, label=r'$x_1(t)=\sin(t)$')
# Plot original x1(t)

axs[0, 0].set_title('Original Signal $x_1(t)$')
# Set title for original x1 plot

axs[0, 0].set_ylabel('Amplitude')
# Label y-axis

axs[0, 0].legend()
# Show legend

axs[0, 0].grid(True, alpha=0.3)
# Enable grid

axs[1, 0].plot(t, x1_scaled, '--', color='blue', linewidth=2, label=r'$\alpha x_1(t)$')
# Plot scaled version of x1(t)

axs[1, 0].set_title('Scaled Signal $\\alpha x_1(t)$')
# Set title for scaled x1 plot

axs[1, 0].set_xlabel('Time $t$')
# Label x-axis

axs[1, 0].set_ylabel('Amplitude')
# Label y-axis

axs[1, 0].legend()
# Show legend

axs[1, 0].grid(True, alpha=0.3)
# Enable grid

# --------------------------------------------------
# Column 2: x2(t)
# --------------------------------------------------
axs[0, 1].plot(t, x2, color='green', linewidth=2, label=r'$x_2(t)=\cos(t)$')
# Plot original x2(t)

axs[0, 1].set_title('Original Signal $x_2(t)$')
# Set title for original x2 plot

axs[0, 1].legend()
# Show legend

axs[0, 1].grid(True, alpha=0.3)
# Enable grid

axs[1, 1].plot(t, x2_scaled, '--', color='green', linewidth=2, label=r'$\alpha x_2(t)$')
# Plot scaled version of x2(t)

axs[1, 1].set_title('Scaled Signal $\\alpha x_2(t)$')
# Set title for scaled x2 plot

axs[1, 1].set_xlabel('Time $t$')
# Label x-axis

axs[1, 1].legend()
# Show legend

axs[1, 1].grid(True, alpha=0.3)
# Enable grid

# --------------------------------------------------
# Final layout adjustment
# --------------------------------------------------
plt.tight_layout()
# Optimize spacing between subplots

plt.show()
# Display the figure
```

## Define the system and apply that to the original and scaled signal

```python
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------
# Apply the system to original and scaled inputs
# --------------------------------------------------

y1 = np.zeros_like(x1)                 # System output for input x1
y2 = np.zeros_like(x2)                 # System output for input x2

y1_scaled = np.zeros_like(x1_scaled)   # System output for scaled input x1
y2_scaled = np.zeros_like(x2_scaled)   # System output for scaled input x2

for i in range(len(t)):                # Iterate over all time samples
    if t[i] >= 0:                      # System is active only for t â‰¥ 0
        y1[i] = x1[i]                  # y1(t) = x1(t)
        y2[i] = x2[i]                  # y2(t) = x2(t)
        y1_scaled[i] = x1_scaled[i]    # y1_scaled(t) = Î± x1(t)
        y2_scaled[i] = x2_scaled[i]    # y2_scaled(t) = Î± x2(t)

# --------------------------------------------------
# Compute homogeneity ratio:  T{Î±x(t)} / (Î± T{x(t)})
# --------------------------------------------------

eps = 1e-9                             # Small value to avoid division by zero
ratio_x1 = y1_scaled / (alpha * y1 + eps)  # Homogeneity ratio for x1
ratio_x2 = y2_scaled / (alpha * y2 + eps)  # Homogeneity ratio for x2

# --------------------------------------------------
# Scientific plotting for homogeneity verification
# --------------------------------------------------

fig, axs = plt.subplots(3, 2, figsize=(15, 12), sharex=True)  # Create a 3Ã—2 subplot layout

# ==================================================
# x1(t)
# ==================================================

axs[0, 0].plot(t, x1, label=r'$x_1(t)$', linewidth=2)             # Plot original input x1
axs[0, 0].plot(t, x1_scaled, '--', label=r'$\alpha x_1(t)$')      # Plot scaled input x1
axs[0, 0].set_title('Input Signals for $x_1(t)$')                  # Set plot title
axs[0, 0].set_ylabel('Amplitude')                                  # Label y-axis
axs[0, 0].legend()                                                  # Show legend
axs[0, 0].grid(True, alpha=0.3)                                     # Enable grid

axs[1, 0].plot(t, y1, label=r'$T\{x_1(t)\}$', linewidth=2)         # Plot system output for x1
axs[1, 0].plot(t, y1_scaled, '--', label=r'$T\{\alpha x_1(t)\}$')  # Plot output for scaled input
axs[1, 0].set_title('System Output for $x_1(t)$')                  # Set plot title
axs[1, 0].set_ylabel('Amplitude')                                  # Label y-axis
axs[1, 0].legend()                                                  # Show legend
axs[1, 0].grid(True, alpha=0.3)                                     # Enable grid

axs[2, 0].plot(t, ratio_x1, color='black', linewidth=2)           # Plot homogeneity ratio for x1
axs[2, 0].axhline(1, color='red', linestyle='--')                  # Reference line y = 1
axs[2, 0].set_title(r'$\frac{T\{\alpha x_1(t)\}}{\alpha T\{x_1(t)\}}$')  # Ratio title
axs[2, 0].set_xlabel('Time $t$')                                    # Label x-axis
axs[2, 0].set_ylabel('Ratio')                                       # Label y-axis
axs[2, 0].grid(True, alpha=0.3)                                     # Enable grid

# ==================================================
# x2(t)
# ==================================================

axs[0, 1].plot(t, x2, label=r'$x_2(t)$', linewidth=2)             # Plot original input x2
axs[0, 1].plot(t, x2_scaled, '--', label=r'$\alpha x_2(t)$')      # Plot scaled input x2
axs[0, 1].set_title('Input Signals for $x_2(t)$')                  # Set plot title
axs[0, 1].legend()                                                  # Show legend
axs[0, 1].grid(True, alpha=0.3)                                     # Enable grid

axs[1, 1].plot(t, y2, label=r'$T\{x_2(t)\}$', linewidth=2)         # Plot system output for x2
axs[1, 1].plot(t, y2_scaled, '--', label=r'$T\{\alpha x_2(t)\}$')  # Plot output for scaled input
axs[1, 1].set_title('System Output for $x_2(t)$')                  # Set plot title
axs[1, 1].legend()                                                  # Show legend
axs[1, 1].grid(True, alpha=0.3)                                     # Enable grid

axs[2, 1].plot(t, ratio_x2, color='black', linewidth=2)           # Plot homogeneity ratio for x2
axs[2, 1].axhline(1, color='red', linestyle='--')                  # Reference line y = 1
axs[2, 1].set_title(r'$\frac{T\{\alpha x_2(t)\}}{\alpha T\{x_2(t)\}}$')  # Ratio title
axs[2, 1].set_xlabel('Time $t$')                                    # Label x-axis
axs[2, 1].set_ylabel('Ratio')                                       # Label y-axis
axs[2, 1].grid(True, alpha=0.3)                                     # Enable grid

plt.tight_layout()                                                  # Adjust subplot spacing
plt.show()                                                          # Display the final figure
```

## Define the System for the additivity test

```python
# --------------------------------------------------
# Apply system for additivity test
# --------------------------------------------------

x_sum = x1 + x2
# Compute the sum of the two input signals: x1(t) + x2(t)

y1 = np.zeros_like(x1)
# Initialize the system output corresponding to x1(t)

y2 = np.zeros_like(x2)
# Initialize the system output corresponding to x2(t)

y_sum = np.zeros_like(x_sum)
# Initialize the system output corresponding to x1(t) + x2(t)

y_sum_expected = np.zeros_like(x_sum)
# Initialize expected output based on additivity: y1(t) + y2(t)

for i in range(len(t)):
    # Loop over all time samples

    if t[i] >= 0:
        # Apply the system condition u(t): system is active only for t >= 0

        y1[i] = x1[i]
        # System output for x1(t): y1(t) = x1(t) for t >= 0

        y2[i] = x2[i]
        # System output for x2(t): y2(t) = x2(t) for t >= 0

        y_sum[i] = x_sum[i]
        # System output for x1(t) + x2(t)

        y_sum_expected[i] = y1[i] + y2[i]
        # Expected output based on additivity: y1(t) + y2(t)
```

```python
# --------------------------------------------------
# Plot: Additivity
# --------------------------------------------------

fig, axs = plt.subplots(2, 2, figsize=(14, 9), sharex=True)
# Create a 2x2 subplot layout for additivity analysis

# --------------------------------------------------
# Top Left: Individual Inputs
# --------------------------------------------------
axs[0, 0].plot(t, x1, color='blue', linewidth=2, label=r'$x_1(t)$')
# Plot first input signal x1(t)

axs[0, 0].plot(t, x2, color='green', linewidth=2, label=r'$x_2(t)$')
# Plot second input signal x2(t)

axs[0, 0].set_title('Individual Inputs')
# Set title for individual inputs plot

axs[0, 0].set_ylabel('Amplitude')
# Label y-axis

axs[0, 0].legend()
# Show legend

axs[0, 0].grid(True, alpha=0.3)
# Enable grid with transparency

# --------------------------------------------------
# Bottom Left: Individual Outputs
# --------------------------------------------------
axs[1, 0].plot(t, y1, color='blue', linewidth=2, label=r'$T\{x_1(t)\}$')
# Plot system output corresponding to x1(t)

axs[1, 0].plot(t, y2, color='green', linewidth=2, label=r'$T\{x_2(t)\}$')
# Plot system output corresponding to x2(t)

axs[1, 0].set_title('System Outputs (Individual)')
# Set title for individual outputs

axs[1, 0].set_xlabel('Time $t$')
# Label x-axis

axs[1, 0].set_ylabel('Amplitude')
# Label y-axis

axs[1, 0].legend()
# Show legend

axs[1, 0].grid(True, alpha=0.3)
# Enable grid

# --------------------------------------------------
# Top Right: Summed Input
# --------------------------------------------------
axs[0, 1].plot(
    t, x_sum, '--', color='purple', linewidth=2,
    label=r'$x_1(t) + x_2(t)$'
)
# Plot the summed input signal

axs[0, 1].set_title('Summed Input')
# Set title for summed input

axs[0, 1].legend()
# Show legend

axs[0, 1].grid(True, alpha=0.3)
# Enable grid

# --------------------------------------------------
# Bottom Right: Additivity Verification
# --------------------------------------------------
axs[1, 1].plot(
    t, y_sum, color='red', linewidth=2,
    label=r'$T\{x_1(t)+x_2(t)\}$'
)
# Plot system output for summed input

axs[1, 1].plot(
    t, y_sum_expected, '--', color='black', linewidth=2,
    label=r'$T\{x_1(t)\}+T\{x_2(t)\}$'
)
# Plot expected output based on additivity property

axs[1, 1].set_title('Additivity Verification')
# Set title for additivity check

axs[1, 1].set_xlabel('Time $t$')
# Labe x-axis

axs[1, 1].legend()
# Show legend

axs[1, 1].grid(True, alpha=0.3)
# Enable grid

# --------------------------------------------------
# Final layout adjustment
# --------------------------------------------------
plt.tight_layout()
# Optimize spacing between subplots

plt.show()
# Display the figure
```

## Final Analysis and Conclusion

The system under analysis is mathematically defined by the operation of multiplying the input signal by the unit step function:

$$
y(t) = x(t)u(t)
$$

We examine the system properties based on standard definitions from signals and systems theory.

---

### Linearity Test

#### **Homogeneity**
Applying a scalar $a$ to the input signal yields:
$$
T\{a x(t)\} = (a x(t))u(t) = a[x(t)u(t)] = aT\{x(t)\}.
$$
Thus, scaling the input results in a proportional scaling of the output.  
The homogeneity property **holds**, and the system is **homogeneous**.

#### **Additivity**
For two arbitrary input signals $x_1(t)$ and $x_2(t)$:
$$
T\{x_1(t)+x_2(t)\}
= (x_1(t)+x_2(t))u(t)
= x_1(t)u(t) + x_2(t)u(t)
= T\{x_1(t)\} + T\{x_2(t)\}.
$$
The output corresponding to the sum of inputs equals the sum of the individual outputs.  
The additivity property **holds**, and the system is **additive**.

---

### Conclusion on Linearity

Since both homogeneity and additivity are satisfied, the system is rigorously classified as a **Linear System**.
