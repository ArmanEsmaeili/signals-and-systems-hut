### problem systement
Determine if the system with the input_output relationship
\[
    y[n]=n^2 x[n]
\]
is linear.

 ### problem Assumptions 

_The system is a discrete-time system.
_input signal is denoted by \(x[n]).
_Output signal is defined as \(y[n]=n^2 x[n]\).
_Linearity is tested using the superposition principle.
_Two arbitrary input signals \(x_1[n]\) and \(x_2[n]\).
_Constants \(a\) and \(b\) are real scalars.

### Linearity Test 
A system is linear if it satisfies the superposition principle:
\[
    T\{a x_1[n] + b x_2[n]\} = aT\{x_1[n]\} + bT\{x_[n]\}
\]
We apply this definition to the given system.

```python
import sympy as sp

# Define symbols
n = sp.symbols('n', integer=True)
a, b = sp.symbols('a b', real=True)
x_1 = sp.Function('x_1')
x_2 = sp.Function('x_2')

# Define system
def system (x):
    return n**2*x

# Left-hand side
lhs = system(a*x_1(n) + b*x_2(n))

# Right-hand side
rhs = a*system(x_1(n)) + b*system(x_2(n))

sp.simplify(lhs - rhs)
```

```python
#Numerical test (example)
x1_num=n
x2_num=1

a_val=2
b_val=-1

lhs_num =system(a_val*x1_num + b_val*x2_num)
rhs_num =a_val*system(x1_num) +b_val*system(x2_num)

sp.simplify(lhs_num - rhs_num)
```

 ### Final Analysis
 The system satisfies both additivity and homogeneity properties.
 Therefore, the superposition principle holds.

 Although the system depends on time index \(n\), this does not affect linearity.

    The system is linear.
    The system is time_varying.
