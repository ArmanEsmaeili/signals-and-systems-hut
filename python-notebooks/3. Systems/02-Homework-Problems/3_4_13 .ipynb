### === Cell 4 - Solution Section ===

#Step 1:

Stability Criterion

A discrete-time system is **BIBO stable** if every bounded input produces a bounded output.

For LTI systems, this is equivalent to the condition that the **impulse response** is absolutely summable.


#Step 2:

 Impulse Response DerivationThe impulse response, denoted as $h[n]$, is the system's output when the input is a discrete-time impulse function. To find it, we substitute the input $x[n]$ with the Kronecker delta function $\delta[n]$:$$x[n] = \delta[n]$$Substituting this into the given system difference equation:$$y[n] = x[n] + x[n-1]$$We obtain the impulse response $h[n]$ as:$$h[n] = \delta[n] + \delta[n-1]$$This results in the following sequence values:$h[n] = 1$ for $n = 0, 1$$h[n] = 0$ otherwise= delta[n] + delta[n-1]]

#Step 3:


```python
import sympy as sym
from IPython.display import display, Math

n = sym.symbols('n', integer=True)

# Define discrete-time impulse response
h = sym.KroneckerDelta(n, 0) + sym.KroneckerDelta(n-1, 0)

display(Math(r"h[n] = " + sym.latex(h)))
```

#Step 4:
Stability Analysis (BIBO Test)A discrete-time Linear Time-Invariant (LTI) system is considered Bounded-Input Bounded-Output (BIBO) stable if and only if its impulse response $h[n]$ is absolutely summable.The mathematical condition for stability is defined as:$$\sum_{n=-\infty}^{\infty} |h[n]| < \infty$$Calculation:Given our impulse response $h[n] = \delta[n] + \delta[n-1]$, we calculate the sum as follows:$$\sum_{n=-\infty}^{\infty} |h[n]| = |h[0]| + |h[1]|$$$$\sum_{n=-\infty}^{\infty} |h[n]| = |1| + |1| = 2$$Conclusion:Since the sum is a finite value ($2 < \infty$), the system satisfies the BIBO stability criterion. Therefore, the system is Stable.

```python
# Absolute value of impulse response
abs_h = sym.Abs(h)

# Compute summation
stability_sum = sym.summation(abs_h, (n, -sym.oo, sym.oo))

display(Math(r"\sum_{n=-\infty}^{\infty} |h[n]| = " + sym.latex(stability_sum)))
```

Since the impulse response consists of only two non-zero values, we can evaluate its magnitude at each discrete time step. Given:$$h[n] = \delta[n] + \delta[n-1]$$The absolute value $|h[n]|$ is determined as:For $n=0$: $|h[0]| = |1| = 1$For $n=1$: $|h[1]| = |1| = 1$For $n \neq 0, 1$: $|h[n]| = 0$values at n=0 and n=1]Consequently, the magnitude of the impulse response can be summarized as:$$|h[n]| =
\begin{cases}
1 & \text{if } n = 0, 1 \\
0 & \text{otherwise}
\end{cases}$$This confirms that the impulse response is finite in duration and magnitude, which is a key observation for the upcoming stability test.

```python
import numpy as np

# Discrete impulse response values
h_values = np.array([1, 1])

# Absolute sum
np.sum(np.abs(h_values))
```
