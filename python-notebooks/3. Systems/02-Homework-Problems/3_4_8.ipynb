# Problem Assumptions & Method
"""

                         بررسی خطی بودن سیستم‌ها

سیستم‌های زمان گسسته زیر مورد بررسی قرار می‌گیرند:

                 1) y[n] = x[n^2]
                 2) y[n] = x^2[n]

برای تعیین خطی بودن سیستم‌ها از اصل ابرموقعیت استفاده می‌کنیم
این اصل شامل دو ویژگی اصلی است: همگنی و جمع‌پذیری

تعریف خطی بودن سیستم:
سیستم خطی است اگر برای هر دو ورودی دلخواه x1[n] و x2[n] و هر دو ضریب دلخواه a و b،
رابطه زیر برقرار باشد:

                 T(a * x1[n] + b * x2[n]) = a * T(x1[n]) + b * T(x2[n])

توضیح مفاهیم:
   همگنی: اگر ورودی سیستم در یک عدد ثابت ضرب شود، خروجی نیز در همان عدد ضرب می‌شود
   جمع‌پذیری: اگر دو ورودی با هم جمع شوند، خروجی برابر با جمع خروجی‌های جداگانه است

نتیجه‌گیری:
سیستمی که هم همگنی و هم جمع‌پذیری داشته باشد، سیستم خطی است.
این ویژگی‌ها برای بررسی و تحلیل خطی بودن سیستم‌های داده شده استفاده می‌شوند.

```python
## Define Symbols 
import sympy as sp

n = sp.symbols('n', integer=True) ##نماد ریاضی از نوع عدد صحیح چون زمان گسسته است
a, b = sp.symbols('a b')## ضرایب دلخواه 
## سیگنال های ورودی
x1 = sp.Function('x1')(n) 
x2 = sp.Function('x2')(n)
```

## System 1: \( y[n] = x[n^2] \)

We test the superposition principle.


```python
"""
تابع subs در SymPy برای جایگزینی یک متغیر یا یک عبارت در یک عبارت سمبولیک با مقدار یا عبارت دیگر استفاده می‌شود
با استفاده از این تابع می‌توان مقادیر متغیرها را تغییر داد یا جایگزین کرد بدون تغییر اصل عبارت
مثال: x.subs(n, n**2) یعنی در عبارت x، همه nها با n^2 جایگزین می‌شوند
این تابع در شبیه‌سازی سیستم‌های گسسته، اعمال تغییرات روی ورودی‌ها و بررسی ویژگی‌هایی مانند خطی بودن بسیار کاربردی است.
"""

# ترکیب خطی روی خروجی ها
lhs = a * x1.subs(n, n**2) + b * x2.subs(n, n**2)
# ترکیب خطی روی ورودی ها
rhs = (a * x1 + b * x2).subs(n, n**2)
# اگر نتیجه صفر شود سیستم خطی است
sp.simplify(lhs - rhs)
```

### System 2: \( y[n] = x^2[n] \)

Again we test the superposition principle.


```python
lhs = a * (x1**2) + b * (x2**2)
rhs = (a * x1 + b * x2)**2

sp.expand(lhs - rhs)

## جواب غیر 0 هست بنابر این ترکیب خطی نیست
```

## Final Analysis



### سیستم اول: `y[n] = x[n^2]`
- بررسی همگنی و جمع‌پذیری:
  - ترکیب خطی روی خروجی‌ها: `a * x1[n^2] + b * x2[n^2]`
  - خروجی سیستم برای ورودی ترکیبی: `(a * x1 + b * x2)[n^2]`
- نتیجه: برابر هستند  
  **سیستم خطی است**  
  *توضیح:* جایگزینی `n → n^2` جمله غیرخطی ایجاد نمی‌کند و همگنی و جمع‌پذیری برقرار است

### سیستم دوم: `y[n] = x^2[n]`
- بررسی همگنی و جمع‌پذیری:
  - ترکیب خطی روی خروجی‌ها: `a * x1^2[n] + b * x2^2[n]`
  - خروجی سیستم برای ورودی ترکیبی: `(a * x1 + b * x2)^2[n]`
- نتیجه: برابر نیست  
  **سیستم غیرخطی است**  
  *توضیح:* توان گرفتن باعث ایجاد جمله‌های ضربی بین ورودی‌ها می‌شود و همگنی و جمع‌پذیری برقرار نیست
