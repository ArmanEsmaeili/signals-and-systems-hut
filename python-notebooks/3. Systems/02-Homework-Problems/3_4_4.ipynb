## Problem Assumptions

- The input signal is \( x(t) \) (any general signal).
- The output signal is defined as \( y(t) = \cos(x(t)) \).
- The system transforms the input signal to the output signal **instantaneously**, without using past or future values.
- Time is continuous, but for plotting, it will be sampled.
- No integral, derivative, or convolution is involved in this system.

### Step 1: Define the input signal and time axis

We define a range of time instants and compute the input signal values.

```python
import numpy as np
import matplotlib.pyplot as plt

# Time axis
dt = 1.2
t = np.arange(0, 10 + dt, dt)

# Example input signal (you can choose any for demonstration)
x = np.sin(2 * np.pi * 0.2 * t)
```

### Step 2: Compute output signal

The system output at each time instant is obtained by applying the cosine function
to the input signal at the same time instant.

This shows that the system does not store or use any information from past or future times.

```python
# Output signal
y = np.cos(x)
```

### Plot Input Signal

In this step, we plot the input signal \(x(t)\) over the selected time range.  
This visualization helps us see how the signal behaves before applying the system transformation.

- `plt.figure(figsize=(10,6))` sets the size of the plot.  
- `plt.subplot(2,1,1)` creates the top subplot (for the input signal).  
- `plt.plot(t, x)` draws the input signal versus time.  
- `plt.title("Input Signal x(t)")` gives the plot a title.  
- `plt.grid()` adds a grid for better readability.

```python
plt.figure(figsize=(10,6))

plt.subplot(2,1,1)
plt.plot(t, x)
plt.title("Input Signal x(t)")
plt.grid()
```

### Plot Output Signal

Here we plot the output signal \(y(t) = \cos(x(t))\) in the bottom subplot.  

- `plt.subplot(2,1,2)` creates the bottom subplot (for the output).  
- `plt.plot(t, y)` draws the output signal versus time.  
- `plt.title("Output Signal y(t) = cos(x(t))")` gives the plot a descriptive title.  
- `plt.xlabel("Time")` labels the horizontal axis.  
- `plt.grid()` adds a grid to make the plot easier to read.  
- `plt.tight_layout()` ensures the two plots do not overlap.  
- `plt.show()` displays the figure.

```python
plt.subplot(2,1,2)
plt.plot(t,y)
plt.title("Output Signal y(t) = cos(x(t))")
plt.xlabel("Time")
plt.grid()

plt.tight_layout()
plt.show()
```

## Analysis: System Memory

The output at each time instant depends only on the input at the same time instant.
The system does not use any past or future input values to compute the output.

Therefore, the system **has no memory**.
