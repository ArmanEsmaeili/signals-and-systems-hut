##  Problem Assumptions

- Input signal: \( x(t) = e^{-t}u(t-1) \)  
- System \( S_1 \) response: \( y_1(t) = t e^{-t} u(t) \)  
- System \( S_2 \) response: \( y_2(t) = t e^{-t} u(t-2) \)  
- Task: Determine whether each system is causal.  
- Reminder: A system is causal if its output at time \( t_0 \) depends only on input values at \( t \leq t_0 \).

##  Step 1: Define signals

We start by defining the input signal and the given system responses:

- Input: \( x(t) = e^{-t}u(t-1) \)  
- System \( S_1 \) output: \( y_1(t) = t e^{-t} u(t) \)  
- System \( S_2 \) output: \( y_2(t) = t e^{-t} u(t-2) \)  

We will use Python (SymPy) to represent these signals and check causality.

```python
import numpy as np
import matplotlib.pyplot as plt

def u(t):
    return np.where(t >= 0, 1.0, 0.0)

def x(t):
    return np.exp(-t) * u(t - 1)

def y1(t):
    return t * np.exp(-t) * u(t)

def y2(t):
    return t * np.exp(-t) * u(t - 2)

t = np.linspace(-1, 10, 1000)
```

<h3>Signal Definitions</h3>
<p>All signals are defined using exponential decay multiplied by unit step functions.</p>

##  Step 2: Check causality

- A system is causal if its output at time \( t_0 \) depends only on input values at \( t \leq t_0 \).  
- For \( S_1 \): Output is \( y_1(t) = t e^{-t} u(t) \). This output starts at \( t=0 \).  
- For \( S_2 \): Output is \( y_2(t) = t e^{-t} u(t-2) \). This output starts at \( t=2 \).  
- We will check whether each output depends only on past or present input values.

```python
plt.figure(figsize=(9,5))
plt.plot(t, x(t), label='x(t) = e^{-t} u(t-1)', color='green', linewidth=2)
plt.plot(t, y1(t), label='y₁(t) = t e^{-t} u(t)', color='blue', linewidth=2)
plt.title("Comparison: x(t) vs y₁(t)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```

<h3>X vs Y₁</h3>
<p>x(t) starts at t = 1, while y₁(t) starts at t = 0.  
Since the output is active before the input, system S₁ is non-causal.</p>

```python
plt.figure(figsize=(9,5))
plt.plot(t, x(t), label='x(t) = e^{-t} u(t-1)', color='green', linewidth=2)
plt.plot(t, y2(t), label='y₂(t) = t e^{-t} u(t-2)', color='red', linewidth=2)
plt.title("Comparison: x(t) vs y₂(t)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```

<h3>X vs Y₂</h3>
<p>x(t) starts at t = 1, while y₂(t) starts at t = 2.  
Since the output begins after the input, system S₂ is causal.</p>

<h3>Result Analysis</h3>
<p>S₁ is non-causal because its output starts before the input.  
S₂ is causal because its output starts after the input.</p>

<h3>Mathematical Solution</h3>
<p>x(t) = e^{-t} u(t-1)<br>
y₁(t) = t e^{-t} u(t)<br>
y₂(t) = t e^{-t} u(t-2)</p>

<h3>Key Characteristics</h3>
<ul>
  <li>Exponential decay</li>
  <li>Time-shifted activation</li>
  <li>Causality depends on input timing</li>
</ul>

<h3>Physical Interpretation</h3>
<p>S₁ reacts before input arrives → non-causal.  
S₂ reacts after input arrives → causal.</p>
