### === Cell 3 - Assumptions & Method ===






In this problem, a discrete-time system composed of three interconnected
subsystems is considered.

- The system operates in **discrete time**, and all signals are defined as
  sequences indexed by the integer variable \( n \).

- The input and output signals of the overall system are denoted by
  \( x[n] \) and \( y[n] \), respectively.

- The impulse responses of the subsystems are given as finite-length sequences:

  \
  h_1[n] =
  \begin{cases}
  2,\;3,\;4 & 0 \leq n \leq 2 \\
  0 & \text{otherwise}
  \end{cases}

  \
  h_2[n] =
  \begin{cases}
  -1,\;3,\;1 & 0 \leq n \leq 2 \\
  0 & \text{otherwise}
  \end{cases}

  \
  h_3[n] =
  \begin{cases}
  1,\;1,\;-1 & 0 \leq n \leq 2 \\
  0 & \text{otherwise}
  \end{cases}



- Subsystems \( S_1 \) and \( S_2 \) are connected in **parallel**, and their
  outputs are added.

- The resulting equivalent system is connected in **cascade** with subsystem
  \( S_3 \).

- All subsystems are assumed to be **Linear and Time-Invariant (LTI)**.

- For part (b) of the problem, the input signal is defined as:
  \[
  x[n] = u[n] - u[n-2]
  \]

- The required operations in this problem include addition of impulse responses
  and discrete-time convolution.


### === Cell 4 - Solution Section ===

Import Libraries

```python
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

Define Time Variables

```python
n = sym.symbols('n', integer=True)
k = sym.symbols('k', integer=True)
```

```python
def print_latex(name, expr):
    display(Math(name + "=" + sym.latex(expr)))
```

Define impulse response h1[n] for n = 0,1,2


```python
h1 = [2, 3, 4]
plt.stem(h1)
plt.title("Impulse Response h[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

Define Impulse Response h[n]

```python
h2 = [-1, 3, 1]

plt.stem(h2)
plt.title("Impulse Response h[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

parallel connection of h1[n] and h2[n]

```python
h12 = np.array(h1) + np.array(h2)
print("h12 =", h12)
plt.stem(h12)
plt.title("h[n] = h[n] + h[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

Define Impulse Response h[n]

```python
h3 = [1, 1, -1]

plt.stem(h3)
plt.title("Impulse Response h[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

Series connection using discrete-time convolution


```python
h = np.convolve(h12, h3)
print("Overall impulse response h =", h)
plt.stem(h)
plt.title("Overall Impulse Response h[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

Define input signal x[n] = u[n] - u[n-2]


```python
x = [1, 1]
print("x =", x)

plt.stem([1, 1, 0, 0, 0])
plt.title("Input Signal x[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

Compute output signal using convolution


```python
y = np.convolve(x, h)
print("y =", y)

plt.stem(y)
plt.title("Output Signal y[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
