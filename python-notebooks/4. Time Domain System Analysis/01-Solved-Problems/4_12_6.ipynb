# === Cell 3 - Assumptions & method ===

<h2>Assumptions &amp; Method</h2>

<h3>Assumptions</h3>
<ol>
  <li>
    <b>Continuous-Time Signal Approximation</b><br />
    The signals <span>x(t)</span>, <span>h(t)</span>, and <span>y(t)</span> are originally
    continuous-time signals. In this implementation, they are approximated using discrete-time
    samples with a uniform sampling interval <span>&Delta;t = 0.1</span>.
  </li>

  <li>
    <b>Linearity and Time Invariance</b><br />
    The system is assumed to be <b>linear and time-invariant (LTI)</b>, allowing the use of the
    convolution operation to compute the system output.
  </li>

  <li>
    <b>Causality</b><br />
    Both the input signal <span>x(t)</span> and the impulse response <span>h(t)</span> are assumed
    to be causal, meaning they are zero for <span>t &lt; 0</span>. Consequently, the output signal
    <span>y(t)</span> is also causal.
  </li>

  <li>
    <b>Finite Signal Duration</b><br />
    The input signal <span>x(t)</span> is of finite duration, while the impulse response
    <span>h(t)</span> decays exponentially and is effectively time-limited for practical
    computation.
  </li>

  <li>
    <b>Numerical Convolution Approximation</b><br />
    The continuous-time convolution integral is approximated using a discrete summation. A scaling
    factor equal to the sampling interval is applied to preserve the correct amplitude.
  </li>
</ol>

# === Cell 4 - Solution Section ===

<h3>Time Discretization</h3> <p>A uniform discrete-time grid is created with step size 0.1 to approximate continuous-time signals. This resolution controls numerical accuracy in convolution.</p>

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 10.1, 0.1)
x = np.piecewise(t, 
                 [t <= 2, (t > 2) & (t <= 4)], 
                 [lambda t: t, lambda t: 4 - t, 0])

plt.figure(figsize=(8,4))
plt.plot(t, x)
plt.title("Input Signal x(t)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
h = t * np.exp(-t)

plt.figure(figsize=(8,4))
plt.plot(t, h)
plt.title("Impulse Response h(t)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

<h3>Output Signal Visualization</h3> <p>This section plots the system output obtained from discrete convolution. The time axis is adjusted to match the length of the convolution result.</p>

```python
y = np.convolve(x, h) * 0.1
t_y = np.arange(len(y)) * 0.1

plt.figure(figsize=(8,4))
plt.plot(t_y, y)
plt.title("Output Signal y(t) = x(t) * h(t)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

  </li>
  <li>

<h3>Convolution Result Interpretation</h3>
<ul>
  <li>
    The output signal <span>y(t)</span> begins at zero and gradually increases as the triangular

  </li>
  <li>

    overlapping portions of <span>x(t)</span> and <span>h(t)</span> contribute to the output.
  </li>
  <li>
    After the input signal ends at <span>t = 4</span>, the output does not abruptly drop to zero;

</ul>

<h3>Effect of the Impulse Response</h3>
<ul>

  <li>

    of the input before the exponential decay dominates.
  </li>
</ul>
