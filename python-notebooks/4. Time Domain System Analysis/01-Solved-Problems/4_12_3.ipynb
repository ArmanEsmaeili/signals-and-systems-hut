### Problem Statement
The input signal
\[
    x(t)=u(t+2) - u(t+3)
\]
is applied to an LTI system with impulse response
\[
h(t)=e^{-3t}u(t)
\]

compute and plot the output signal of the system.

### Problem Assumptions

- The system is linear and time-invariant (LTI).
- The input signal is a rectangular pulse defined between \(t=-2\) and \(t=3\).
- The impulse response is a causal exponential signal.
- Continuous-time convolution is approximated using numerical methods.
- Sampling interval is assumed as \(t = 0.1\).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=0.1 
#Time axis for input
t_x=np.arange(-2 ,10,dt )
#Input signal x(t)
x=np.where((t_x >= -2) & (t_x <= 3),1,0)
#Time axis for impulse response
t_h =np.arange (0 ,10 ,dt)
#Impulse response h(t)
h= np.exp(-3 * t_h)
```

```python
y = np.convolve (x ,h)*dt

#Time axis for output
t_y =np.arange (0,len(y)) * dt + (t_x[0] +t_h[0])
```

```python
plt.figure()
plt.plot(t_x,x)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Input Signal x(t)")
plt.grid()
plt.show()
```

```python
plt.figure()
plt.plot(t_h,h)
plt.xlabel("t")
plt.ylabel("h(t)")
plt.title("impulse response h(t)")
plt.grid()
plt.show()
```

```python
plt.figure()
plt.plot(t_y,y)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.title("Output Signal y(t)")
plt.grid()
plt.legend(["y(t)"])
plt.show()
```

### Final Analysis

The output signal is obtained by convolving the rectangular input signal
with the causal exponential impulse response.

The resulting signal starts at \(t = -2\) due to the finite support of the input
and decays exponentially due to the nature of the impulse response.

The shape of the output is consistent with the expected theoretical behavior
of an LTI system.
