## Problem Assumptions

- The system is a continuous-time LTI system characterized by its impulse response.
- The impulse response is given as:
  \[
  h(t) = t,\quad 0 \leq t \leq 10
  \]
  and is zero outside this interval.
- The input signal is defined as:
  \[
  x(t) = 0.8^t,\quad 0 \leq t \leq 10
  \]
  and zero elsewhere.
- The output signal is obtained using the continuous-time convolution integral.
- The objective is to compute and plot the system response \( y(t) \).

## Step 1: Definition of Time Axis and Signals

In this step, the time axis is discretized in order to numerically approximate the continuous-time convolution integral. The input signal and impulse response are defined according to the problem statement.

```python
import numpy as np
import matplotlib.pyplot as plt

# Time resolution
dt = 7

# Time axes
t = np.arange(0, 10 + dt, dt)

# Input signal x(t)
x = 0.8 ** t

# Impulse response h(t)
h = t
```

## Step 2: Numerical Implementation of Convolution

In this step, the output signal of the system is computed based on the definition of
continuous-time convolution.

For each time instant t, the output y(t) is obtained by combining past values of the input
signal with the impulse response of the system.

Since the signals are defined only over a finite time interval and are sampled with a
small time step, the convolution integral is approximated using a summation.
The impulse response h(t − τ) is evaluated only when it lies within its defined range.

```python
# Output time axis
t_y = np.arange(0, 20 + dt, dt)

# Initialize output
y = np.zeros(len(t_y))

# Numerical convolution
for i in range(len(t_y)):
    for j in range(len(t)):
        if 0 <= t_y[i] - t[j] <= 10:
            y[i] += x[j] * (t_y[i] - t[j]) * dt
```

## Step 3: Signal Visualization

In this step, the input signal, impulse response, and system output are plotted for visualization and interpretation.

```python
plt.figure(figsize=(10, 7))

plt.subplot(3, 1, 1)
plt.plot(t, x)
plt.title("Input Signal x(t) = 0.8^t")
plt.grid()
```

```python
plt.subplot(3, 1, 2)
plt.plot(t, h)
plt.title("Impulse Response h(t) = t")
plt.grid()
```

```python
plt.subplot(3, 1, 3)
plt.plot(t_y, y)
plt.title("Output Signal y(t) = x(t) * h(t)")
plt.xlabel("Time")
plt.grid()

plt.tight_layout()
plt.show()
```

## Final Analysis

The output signal is obtained by convolving the input signal with the impulse response of the system. The result shows a smooth increasing behavior, which is expected due to the increasing nature of the impulse response. The system response is consistent with the theoretical definition of continuous-time convolution for LTI systems.
