<h2>Assumptions &amp; Method</h2>

<h3>Assumptions</h3>
<ol>
  <li>
    <b>Ideal Dirac Delta (Distribution)</b><br />
    The Dirac delta <span>&delta;(t - t<sub>0</sub>)</span> is treated as an ideal mathematical distribution (generalized function), not a regular function. It is characterized by:
    <ul>
      <li>
        <span>&delta;(t - t<sub>0</sub>) = 0</span> for <span>t &ne; t<sub>0</sub></span>
      </li>
      <li>
        <span>&int;<sub>-&infin;</sub><sup>&infin;</sup> &delta;(t - t<sub>0</sub>) dt = 1</span>
      </li>
      <li>
        <span>&int;<sub>-&infin;</sub><sup>&infin;</sup> f(t) &delta;(t - t<sub>0</sub>) dt = f(t<sub>0</sub>)</span> (sifting property), for a well-behaved function <span>f(t)</span>
      </li>
    </ul>
  </li>

  <li>
    <b>Visualization Assumption</b><br />
    Because the Dirac delta cannot be plotted directly as a finite-valued curve, it is visualized as a single impulse (a vertical arrow) located at <span>t = t<sub>0</sub></span>. The displayed arrow height is chosen only for clarity and does not represent a true amplitude.
  </li>

  <li>
    <b>Finite Time Interval for Plotting</b><br />
    The plot is generated over a finite interval <span>[t<sub>0</sub> - T,\, t<sub>0</sub> + T]</span> to clearly show the impulse position. This plotting interval does not change the mathematical meaning of the delta distribution.
  </li>

  <li>
    <b>Software Tools</b><br />
    <ul>
      <li><b>SymPy</b> is used to define the symbolic expression <span>&delta;(t - t<sub>0</sub>)</span>.</li>
      <li><b>NumPy</b> is used to create a numerical time vector for plotting.</li>
      <li><b>Matplotlib</b> is used to draw the impulse as an arrow and format the graph.</li>
    </ul>
  </li>
</ol>

<h3>Method</h3>
<ol>
  <li>
    <b>Define Symbolic Variables</b><br />
    Create a symbolic time variable <span>t</span> in SymPy.

  <li>

<h3>Symbolic Time Variable</h3> <p>The time variable is defined symbolically to enable analytical manipulation. Restricting it to real values ensures correct physical interpretation.</p>

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def dirac_delta(t0, T=5):
    """
    Generates and visualizes a Dirac delta (impulse) function Î´(t - t0).

    Parameters
    ----------
    t0 : float
        Time location of the impulse.
    T : float, optional
        Time range for visualization (default is 5).

    Returns
    -------
    delta_func : sympy.DiracDelta
        Symbolic representation of the Dirac delta function.
    """

    # -------------------------------------------------
    # 1) Symbolic definition of the Dirac delta function
    # -------------------------------------------------
    # Define a symbolic time variable 't'
    t = sp.symbols('t', real=True)

    # Create a symbolic Dirac delta function centered at t = t0
    delta_func = sp.DiracDelta(t - t0)

    # -------------------------------------------------
    # 2) Time axis for visualization (numerical domain)
    # -------------------------------------------------
    # Create a numerical time vector around t0
    # This is only used for plotting purposes
    time = np.linspace(t0 - T, t0 + T, 1000)

    # -------------------------------------------------
    # 3) Plot impulse representation
    # -------------------------------------------------
    plt.figure(figsize=(8, 4))

    # Draw horizontal time axis (amplitude = 0)
    plt.axhline(0, linewidth=1)

    # Draw an arrow to represent the impulse at t = t0
    # The arrow visually represents infinite amplitude and zero width
    plt.arrow(
        t0, 0,              # Starting point of the arrow (t0, 0)
        0, 1,               # Arrow direction (vertical)
        head_width=0.15,
        head_length=0.1,
        length_includes_head=True,
        color='red'
    )

    # Label the impulse
    plt.text(t0, 1.05, r'$\delta(t - t_0)$', ha='center', fontsize=12)

    # Set plot limits
    plt.xlim(t0 - T, t0 + T)
    plt.ylim(-0.2, 1.3)

    # Axis labels and title
    plt.xlabel('t')
    plt.ylabel('Amplitude')
    plt.title(f'Dirac Delta Function Î´(t âˆ’ {t0})')

    # Enable grid for better visualization
    plt.grid(True)

    # Display the plot
    plt.show()

    # -------------------------------------------------
    # 4) Return symbolic expression
    # -------------------------------------------------
    # Return the symbolic Dirac delta function
    return delta_func
```

<h3>Impulse Visualization Setup</h3> <p>This section initializes a clean plotting canvas and draws a reference time axis. A horizontal baseline is included to emphasize the impulse location.</p>

```python

delta = dirac_delta(t0=2)
delta

# Define a test function
t = sp.symbols('t', real=True)
f = sp.sin(t) + t**2

# Apply sifting property
result = sp.integrate(f * sp.DiracDelta(t - 2), (t, -sp.oo, sp.oo))
result
```
