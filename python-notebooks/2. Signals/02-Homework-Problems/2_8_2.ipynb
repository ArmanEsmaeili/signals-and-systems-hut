````python
import numpy as np
import matplotlib.pyplot as plt
# Time axis explanation:
# Given x(t) = 3e^{-j2t}, the angular frequency is Ï‰ = 2 rad/s
# Therefore, the fundamental period is T = Ï€
# The interval [0, 3Ï€] is chosen to display three full periods of the signal
t = np.linspace(0, 3*np.pi, 1000)
x = 3 * np.exp(-1j * 2 * t)

x_real = np.real(x)
x_imag = np.imag(x)
````

<h3>Plot Initialization</h3> <p>A new figure is created to visualize the signal clearly.</p> <h3>Signal Visualization</h3> <p>The real part of the complex signal is plotted as a function of time.</p> <h3>Plot Annotation</h3> <p>Axis labels and a title are added to identify the plotted quantity.</p> <h3>Grid and Display</h3> <p>A grid improves readability, and the plot is rendered on screen.</p> <h3>Result Analysis</h3> <h3>Mathematical Solution</h3> <p>Re{x(t)} = 3cos(2t)</p> <h3>Key Characteristics</h3> <ul> <li>Sinusoidal waveform</li> <li>Constant amplitude</li> </ul> <h3>Physical Interpretation</h3> <p>The plot shows harmonic oscillation corresponding to the real projection of circular motion.</p>

````python
plt.figure()
plt.plot(t, x_real)
plt.title("Real Part of x(t)")
plt.xlabel("t")
plt.ylabel("Re{x(t)}")
plt.grid(True)
plt.show()
````

<h3>Plot Initialization</h3> <p>A new figure is created for the imaginary component of the signal.</p> <h3>Signal Visualization</h3> <p>The imaginary part of x(t) is plotted versus time.</p> <h3>Plot Annotation</h3> <p>Title and axis labels indicate the plotted component.</p> <h3>Grid and Display</h3> <p>Grid lines are enabled for clarity, and the plot is displayed.</p> <h3>Result Analysis</h3> <h3>Mathematical Solution</h3> <p>Im{x(t)} = âˆ’3sin(2t)</p> <h3>Key Characteristics</h3> <ul> <li>Sinusoidal waveform</li> <li>Phase-shifted 90Â° relative to the real part</li> </ul> <h3>Physical Interpretation</h3> <p>Represents the orthogonal projection of circular motion onto the imaginary axis.</p>

````python
plt.figure()
plt.plot(t, x_imag)
plt.title("Imaginary Part of x(t)")
plt.xlabel("t")
plt.ylabel("Im{x(t)}")
plt.grid(True)
plt.grid(True)
plt.show()
````
