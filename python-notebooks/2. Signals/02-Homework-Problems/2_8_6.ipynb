# === Cell 4 - Solution Section ===

<h3>Symbolic Setup</h3> <p>This section initializes the symbolic variable using SymPy, enforcing real-valued time to ensure correct analytical behavior. Symbolic definitions enable exact mathematical manipulation rather than numerical approximation.</p>

<h3>Piecewise Signal Definition</h3> <p>The signal is defined as a piecewise function over time with linear segments and zero padding outside the active interval. This construction is commonly used to model finite-duration continuous-time signals.</p>


```python
import sympy as sp

t = sp.symbols('t', real=True)

x = sp.Piecewise(
    (t, (t >= 0) & (t <= 2)),
    (4 - t, (t > 2) & (t <= 4)),
    (0, True)
)
```

<h2>Definition</h2>
<p>The code constructs transformed versions of a piecewise signal using symbolic substitutions. It generates time reversal (t→−t), time scaling (t→t/2), and combined shift–scale forms (t→2+4t and t→−2−4t), which change the signal’s timing and support while preserving its shape. These transformations correspond to practical operations like inversion, stretching/compression, and delay/advance in continuous-time signal analysis.</p>

```python
xt = x
xt1 = x.subs(t, -t)
xt2 = x.subs(t, t/2)
xt3 = x.subs(t, 2 + 4*t)
xt4 = x.subs(t, -2 - 4*t)
```

<h3>Physical Interpretation</h3> <p>These plots illustrate how delays, speed changes, and reversals affect real-world signals in systems such as communications and control.</p>

```python
p1 = sp.plot(
    xt,
    (t, -8, 8),
    title="x(t)",
    xlabel="t",
    ylabel="x(t)",
    show=False
)

p1.show()
```

```python
p2 = sp.plot(
    xt1,
    (t, -8, 8),
    title="x(-t)",
    xlabel="t",
    ylabel="x(-t)",
    show=False
)

p2.show()
```

```python
p3 = sp.plot(
    xt2,
    (t, -8, 8),
    title="x(t/2)",
    xlabel="t",
    ylabel="x(t/2)",
    show=False
)
p3.show()
```

```python
p4 = sp.plot(
    xt3,
    (t, -1, 1),
    title="x(2 + 4t)",
    xlabel="t",
    ylabel="x(2 + 4t)",
    show=False
)
p4.show()
```

```python
p5 = sp.plot(
    xt4,
    (t, -2, 0),
    title="x(-2 - 4t)",
    xlabel="t",
    ylabel="x(-2 - 4t)",
    show=False
)
p5.show()
```
