### Assumptions for Question 9

- **Ramp definition:**  
  \( r(t - t0) = 0 \) for \( t < t0 \), and \( r(t - t0) = t - t0 \) for \( t \geq t0 \).

- **Input argument:**  
  The function accepts a single number `t0` which specifies the shift of the ramp.

- **Output:**  
  Returns the values of the shifted ramp over a chosen time vector.

- **Time interval:**  
  Use a symmetric interval around `t0`, e.g. `[t0 - T, t0 + T]` with \( T = 5 \).

- **Sampling step:**  
  A small step size such as `dt = 0.01` ensures smooth plotting.

- **Plotting:**  
  The graph should display `r(t - t0)` versus `t`, with a vertical line at `t = t0` to mark the shift.

- **Behavior:**  
  - Before `t0`: the function is zero.  
  - At `t0`: the function starts at zero.  
  - After `t0`: the function increases linearly with slope 1.

### Methods for Question 9

1. **Define the shifted ramp function**  
   - Implement \( r(t - t0) = (t - t0) * u(t - t_0) \).  
   - Use NumPy for vectorized operations: `(t - t0) * (t >= t0)`.

2. **Build the time vector**  
   - Create a uniform time array with `np.linspace`.  
   - Interval: `[t0 - T, t0 + T]` with step size `dt`.

3. **Compute ramp values**  
   - Apply the ramp function to the time vector.  
   - This produces the array of \( r(t - t_0) \) values.

4. **Plot the result**  
   - Use Matplotlib to plot `r(t - t0)` versus `t`.  
   - Add a vertical dashed line at `t = t0` to mark the shift.  
   - Label axes and add a title for clarity.


# << Cell 4 - Solution Section >>

```python
import numpy as np
import matplotlib.pyplot as plt
# We import NumPy for vectorized math operations and Matplotlib for plotting.
```

 ### step 2  -- Plot function

```python
def ramp_shifted(t, t0):       # We define a function ramp_shifted(t, t0) that computes the shifted ramp ( r(t - t0) ).
    t = np.asarray(t, float)    #Inside the function, we convert t into a NumPy float array to allow elementwise operations.
    return (t - t0) * (t >= t0)   #We return (t - t0) only when t >= t0; otherwise the result is zero.
```

```python
def plot_ramp(t0, T=5.0, dt=0.01):
#We compute the number of samples and create a time vector from t0 - T to t0 + T with step dt.

    n = int(np.ceil(2*T/dt)) + 1               #We evaluate the ramp function on the time vector.
    t = np.linspace(t0 - T, t0 + T, n)
    r = ramp_shifted(t, t0)
    plt.figure(figsize=(7,3.5))               #We open a new figure with a fixed size for plotting.
    plt.plot(t, r, lw=2, label=r"$r(t-t_0)$")   #We plot the ramp values against time with a solid line.
    plt.axvline(t0, color='red', ls='--', label=f"t0={t0}")      #We plot the ramp values against time with a solid line.
    plt.xlabel("t"); plt.ylabel("r(t - t0)")          #We add a dashed red line at t = t0 to mark the shift point.
    plt.title(f"Shifted ramp function, t0={t0}")
    plt.grid(alpha=0.3); plt.legend(); plt.tight_layout()    #We add a light grid, show the legend, and adjust the layout for clarity.
    plt.show()
    return t, r       #We display the plot and return both t and r arrays for reuse.
```

### Step 4 -- Example run

We call `plot_ramp(2.0)` to test the function with `t0 = 2.0`.


```python
plot_ramp(t0=2.0)
```
