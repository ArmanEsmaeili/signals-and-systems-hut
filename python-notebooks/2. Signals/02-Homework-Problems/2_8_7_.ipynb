<h3>Step 1 – Understand the Objective</h3>

The goal is to write a function that takes a real number <b>t0</b> as an input and returns
the shifted unit step signal <b>u(t − t0)</b>. In addition, the function must generate a plot
of <b>u(t − t0)</b> over a suitable time interval.
</p>

<h3>Initialization</h3>

<p>
In this cell, the required libraries are imported and the general setup for the
unit step function analysis is established. A continuous time domain is assumed,
and all computations are performed numerically. The step location parameter
<i>t<sub>0</sub></i> is treated as a real-valued input that determines the shift
of the unit step function <i>u(t − t<sub>0</sub>)</i>. This initialization provides
the foundation for defining the step function, generating the time interval,
and visualizing the results in subsequent steps.
</p>

```python
import numpy as np
import matplotlib.pyplot as plt
```

<h3>Step 2 – Define the Unit Step Function Behavior</h3>
<p>
Use the standard definition of the unit step:
</p>
<ul>
  <li><b>u(t − t0) = 0</b> for <b>t &lt; t0</b></li>
  <li><b>u(t − t0) = 1</b> for <b>t ≥ t0</b></li>
</ul>
<p>
This definition ensures the step occurs exactly at <b>t = t0</b>.
</p>

```python
def unit_step(t, t0):
    return np.where(t >= t0, 1, 0)
```

<h3>Step 3 – Choose a Proper Time Interval for Plotting</h3>
<p>
Select a time range that clearly shows the step transition. A common choice is a symmetric interval around <b>t0</b>, for example:
</p>
<ul>
  <li><b>[t0 − L, t0 + L]</b> where <b>L</b> is a positive value (e.g., 5 or 10)</li>
</ul>
<p>
This guarantees the plot includes values before and after the step.
</p>

```python
def u_shifted(t0):
    return t0
```

<h3>Step 4 – Create a Time Vector</h3>
<p>
Generate a dense set of time samples on the chosen interval. The resolution should be high enough to show a clear transition (e.g., 1000 points).
The output of the function should be computed on this time vector.
</p>

```python
def time_interval(t0, L=5, N=1000):
    t = np.linspace(t0 - L, t0 + L, N)
    return t
```

<h3>Step 5 – Compute u(t − t0) Numerically</h3>
<p>
Evaluate the unit step on the time vector using the definition:
set output values to 0 where <b>t &lt; t0</b>, and set output values to 1 where <b>t ≥ t0</b>.
This produces the discrete-time representation of <b>u(t − t0)</b> needed for plotting.
</p>

```python
def compute_step_signal(t, t0):
    u = unit_step(t, t0)
    return u
```

<h3>Step 6 – Return the Step Signal</h3>
<p>
The function should return the computed step signal values (and typically also return the time vector used).
Returning both <b>t</b> and <b>u(t − t0)</b> is standard practice because it supports reuse and validation outside the plot.
</p>

```python
def step_output(t0):
    t = time_interval(t0)
    u = compute_step_signal(t, t0)
    return t, u
```

<h3>Step 7 – Plot u(t − t0)</h3>
<p>
Plot the step signal against time using a step-style plot (preferred) or a regular line plot.
The plot should include:
</p>
<ul>
  <li>Axis labels (Time and Amplitude)</li>
  <li>A title indicating the shift parameter <b>t0</b></li>
  <li>A grid for readability</li>
  <li>(Optional) A vertical marker at <b>t = t0</b> to highlight the step location</li>
</ul>

```python
def plot_step(t, u, t0):
    plt.figure()
    plt.step(t, u, where='post')
    plt.axvline(t0)
    plt.xlabel("Time (t)")
    plt.ylabel("Amplitude")
    plt.title(f"Unit Step Function u(t - {t0})")
    plt.grid(True)
    plt.show()
```

<h3>Step 8 – Validate with Example Inputs</h3>
<p>
Test the function with a few values of <b>t0</b> (e.g., −2, 0, 3) to confirm:
</p>
<ul>
  <li>The step occurs at the correct location</li>
  <li>The output is 0 before <b>t0</b> and 1 from <b>t0</b> onward</li>
  <li>The selected time interval properly displays the transition</li>
</ul>

```python
t0 = 2
t, u = step_output(t0)
plot_step(t, u, t0)
```
