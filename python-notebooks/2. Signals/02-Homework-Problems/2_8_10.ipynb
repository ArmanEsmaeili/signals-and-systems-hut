# === Cell 4 - Solution Section ===

<h3>Library Initialization</h3> <p>This section imports symbolic, numerical, and plotting libraries required for signal analysis. The combination enables analytical expressions, numerical evaluation, and visualization.</p>

```python
# ==================================================
# Rectangular Function p_T(t) using SymPy
# ==================================================

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------
# 1) Define symbolic variable
# --------------------------------------------------
t = sp.symbols('t', real=True)


```

<h3>Rectangular Pulse Definition</h3> <p>This section defines a symbolic rectangular function using a piecewise formulation. The signal has unit amplitude within a finite symmetric interval and zero elsewhere.</p>

```python
# --------------------------------------------------
# 2) Define rectangular function p_T(t)
# --------------------------------------------------
def rectangular_function(T):
    """
    Returns the symbolic rectangular function p_T(t)
    and plots it over a proper time interval.
    """

    # Piecewise definition of rectangular pulse
    p_T = sp.Piecewise(
        (1, sp.Abs(t) <= T/2),
        (0, True)
    )

    # --------------------------------------------------
    # 3) Plot using SymPy
    # --------------------------------------------------
    sp.plot(
        p_T,
        (t, -T, T),
        xlabel='t',
        ylabel='p_T(t)',
        title=f'Rectangular Function p_T(t) with T = {T}',
        ylim=(-0.2, 1.2),
        show=True
    )

    return p_T

# --------------------------------------------------
# 4) Example usage
# --------------------------------------------------
T_value = 4
pT = rectangular_function(T_value)
pT
```
