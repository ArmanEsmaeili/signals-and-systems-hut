<h3>Assumptions &amp; Methods</h3>

<h4>Assumptions</h4>
<ul>

    The original signal is defined as
    <span style="font-family: 'Times New Roman', serif;">x(t) = t \cos(2\pi t)</span>
    over the interval
    <span style="font-family: 'Times New Roman', serif;">0 \le t \le 5</span>.
  </li>
  <li>
    Outside the given interval, the signal is assumed to be zero (time-limited signal):
    <br/>
    <span style="font-family: 'Times New Roman', serif;">
      x(t)=
      \begin{cases}
        t\cos(2\pi t), &amp; 0 \le t \le 5 \\
        0, &amp; \text{otherwise}
      \end{cases}
    </span>
  </li>
  <li>
    All time transformations (time reversal, scaling, and shifting) are applied to this
    time-limited signal definition.
  </li>
  <li>
    The time axis used for plotting is chosen wide enough to fully capture each transformed
    signal.
  </li>
</ul>

<h4>Methods</h4>
<ul>
  <li>
    A uniformly sampled time vector is created:
    <span style="font-family: 'Times New Roman', serif;">t = \text{linspace}(t_{\min}, t_{\max}, N)</span>.
  </li>
  <li>
    A helper function is used to evaluate the signal using logical masking to enforce the
    interval
    <span style="font-family: 'Times New Roman', serif;">0 \le u \le 5</span>.
  </li>
  <li>
    Each required signal is generated by substituting the appropriate time transformation
    into the original signal definition:
    <ul>

      <li><span style="font-family: 'Times New Roman', serif;">x(1 + 3t)</span>: time scaling and shifting</li>
      <li><span style="font-family: 'Times New Roman', serif;">x(-1 - 3t)</span>: time reversal, scaling, and shifting</li>
    </ul>
  </li>
  <li>
    The resulting signals are plotted using standard time-domain plots with labeled axes and
    appropriate titles for clarity.
  </li>
</ul>

<h3>Step 1 – Environment Setup</h3>
<p>
  We import NumPy for numerical computations and Matplotlib for plotting the signals in the time domain.

```python
import numpy as np
import matplotlib.pyplot as plt
```

<!-- =========================
     Step 2 – Define x(t) with support
     ========================= -->
<h3>Step 2 – Define <em>x(t)</em> with its time support</h3>
<p>
  The signal is given by <strong>x(t) = t cos(2πt)</strong> for <strong>0 ≤ t ≤ 5</strong>.
  Outside this interval, <em>x(t)</em> is assumed to be zero (zero-extension). This assumption is necessary
  to evaluate time transformations such as <em>x(−t)</em> and <em>x(t/5)</em>.

```python
def x(u):
    """
    x(u) = u cos(2πu) for 0<=u<=5, else 0
    """
    u = np.asarray(u)
    y = np.zeros_like(u, dtype=float)
    mask = (u >= 0) & (u <= 5)
    y[mask] = u[mask] * np.cos(2*np.pi*u[mask])
    return y
```

<!-- =========================
     Step 3 – Plot x(t)
     ========================= -->
<h3>Step 3 – Plot <em>x(t)</em></h3>
<p>
  We plot the original signal <em>x(t)</em> over its active interval <strong>0 ≤ t ≤ 5</strong>.

```python
t = np.linspace(0, 5, 2000)
plt.plot(t, x(t))
plt.title("x(t)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

<!-- =========================
     Step 4 – Plot x(-t)
     ========================= -->
<h3>Step 4 – Plot <em>x(−t)</em></h3>
<p>
  Time reversal is applied by replacing <em>t</em> with <em>−t</em>.
  Since <em>x(u)</em> is nonzero only for <strong>0 ≤ u ≤ 5</strong>, <em>x(−t)</em> is nonzero only when
  <em>−t</em> lies in <strong>[0,5]</strong>, i.e., for <strong>−5 ≤ t ≤ 0</strong>.<p>

```python
t = np.linspace(-5, 0, 2000)     # time axis for x(-t)
plt.plot(t, x(-t))
plt.title("x(-t)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

<!-- =========================
     Step 5 – Plot x(t/5)
     ========================= -->
<h3>Step 5 – Plot <em>x(t/5)</em></h3>
<p>
  Time scaling is applied by replacing <em>t</em> with <em>t/5</em>, which expands the signal in time by a factor of 5.
  The signal is nonzero when <em>t/5</em> lies in <strong>[0,5]</strong>, i.e., for <strong>0 ≤ t ≤ 25</strong>.
</p>

```python
t = np.linspace(0, 15, 4000)   # time axis limited to [0, 15]
y = x(t/5)

plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("Time Expansion of x(t/5) over 0 ≤ t ≤ 15")
plt.grid(True)

# Highlight original support of x(t): [0, 5]
plt.axvspan(0, 5, alpha=0.15)

# Original support boundaries
plt.axvline(0, linestyle="--", linewidth=2)
plt.axvline(5, linestyle="--", linewidth=2)

# Expanded support of x(t/5): [0, 25] (partially visible)
plt.axvline(15, linestyle=":", linewidth=2)

# Annotations on the figure
plt.text(0.5, 0.9*np.max(y), "Original support: 0–5", ha="left")
plt.text(6,   0.9*np.max(y), "Expanded support continues beyond 15", ha="left")

plt.show()
```

<!-- =========================
     Step 6 – Plot x(1+3t)
     based on 1+3t in [0,5] => t in [-1/3, 4/3]
     ========================= -->
<h3>Step 6 – Plot <em>x(1 + 3t)</em></h3>
<p>
  This transformation applies a time scaling by 3 and a time shift by +1 in the argument.
  The signal is nonzero when <em>1 + 3t</em> lies in <strong>[0,5]</strong>, which gives
  <strong>−1/3 ≤ t ≤ 4/3</strong>.
</p>

```python
t = np.linspace(-1, 2, 4000)
y = x(1 + 3*t)

plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("x(1 + 3t): Time Scaling + Shift")
plt.grid(True)

# Support boundaries
plt.axvline(-1/3, linestyle="--", linewidth=2)
plt.axvline(4/3,  linestyle="--", linewidth=2)

# Annotations
plt.text(-0.3, 0.9*np.max(y), "Start: t = -1/3", ha="left")
plt.text(0.8,  0.9*np.max(y), "End: t = 4/3",  ha="left")

plt.show()
```

<!-- =========================
     Step 7 – Plot x(-1-3t)
     based on -1-3t in [0,5] => t in [-2, -1/3]
     ========================= -->
<h3>Step 7 – Plot <em>x(−1 − 3t)</em></h3>
<p>
  This transformation includes time reversal and scaling (factor 3) along with a shift.
  The signal is nonzero when <em>−1 − 3t</em> lies in <strong>[0,5]</strong>, resulting in
  <strong>−2 ≤ t ≤ −1/3</strong>.
</p>

```python
t = np.linspace(-2.5, 0.5, 4000)
y = x(-1 - 3*t)

plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.title("x(-1 - 3t): Time Reversal + Scaling + Shift")
plt.grid(True)

# Support boundaries
plt.axvline(-2,   linestyle="--", linewidth=2)
plt.axvline(-1/3, linestyle="--", linewidth=2)

# Annotations
plt.text(-1.9, 0.9*np.max(y), "Start: t = -2",    ha="left")
plt.text(-0.9, 0.9*np.max(y), "End: t = -1/3",  ha="left")

plt.show()
```

<!-- =========================
     Step 8 – Plot all signals in one figure (Optional)
     ========================= -->
<h3>Step 8 – Plot all signals in one figure (Optional)</h3>
<p>
  For better visualization and comparison, all transformed signals are plotted together in a single figure
  using multiple subplots. This representation allows the effects of time reversal, time scaling, and time shifting
  to be observed simultaneously.
</p>

```python
import numpy as np
import matplotlib.pyplot as plt

# Example signal function
def x(t):
    return np.sin(2 * np.pi * t)  # Replace with your actual function

# Create subplots
fig, axs = plt.subplots(5, 1, figsize=(10, 12))

# Time vectors
t1 = np.linspace(0, 5, 2000)
t2 = np.linspace(-5, 0, 2000)
t3 = np.linspace(0, 25, 2500)
t4 = np.linspace(-1/3, 4/3, 2000)
t5 = np.linspace(-2, -1/3, 2000)

# Plot each transformation
axs[0].plot(t1, x(t1)); axs[0].set_title("x(t)")
axs[1].plot(t2, x(-t2)); axs[1].set_title("x(-t)")
axs[2].plot(t3, x(t3/5)); axs[2].set_title("x(t/5)")
axs[3].plot(t4, x(1 + 3*t4)); axs[3].set_title("x(1+3t)")
axs[4].plot(t5, x(-1 - 3*t5)); axs[4].set_title("x(-1-3t)")

# Determine global limits for x and y axes
all_t = np.concatenate([t1, t2, t3, t4, t5])
all_y = np.concatenate([x(t1), x(-t2), x(t3/5), x(1 + 3*t4), x(-1 - 3*t5)])
x_limits = [all_t.min(), all_t.max()]
y_limits = [all_y.min(), all_y.max()]

# Apply same scale to all subplots
for ax in axs:
    ax.grid(True)
    ax.set_xlabel("t")
    ax.set_ylabel("Amplitude")
    ax.set_xlim(x_limits)
    ax.set_ylim(y_limits)

plt.tight_layout()
plt.show()
```
