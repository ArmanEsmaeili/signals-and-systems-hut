# === Cell 3 - Assumptions & method ===

## Assumptions :
- The signal is piecewise linear and defined over the interval
$$ùë°‚àà[0,3].$$


- The signal starts at 0, ramps up linearly from
$ùë°=0$ to,$t=1$ ,stays constant from $t=1$ to $t=2$
, and ramps down linearly to 0 from
$ùë°=2$to$t=3.

- Ramp functions are used to construct the signal. A ramp function is defined as
$ùëü(t)=t , u(t)$ ,where $ u(t)$ is the unit step function.

## methode:
1-Decompose the signal into ramp segments:
- The rising part from $ùë°=0$ to $ùë° =1$ is modeled by $ùëü(ùë°).$
- The flat part from $ùë°=1$ to $ùë°=2$ is achieved by subtracting $ùëü(ùë°‚àí1)$ which cancels the linear growth beyond $ùë°=1$.
- The falling part from $ùë°=2$ to $ùë°=3$
 is modeled by subtracting another ramp $ùëü(ùë°‚àí2)$
, which introduces a negative slope.

2-Final expression:

- $x(t)=r(t)-r(t-1)-r(t-2)$

3-MATLAB implementation:

- Use heaviside(t) to simulate the unit step function.

- Multiply each ramp by its corresponding shifted time and step function.

- Plot the result over a suitable time range.

# === Cell 4 - Solution Section ===

### Cell 1: Imports and time axis
Prepare the environment and time axis for plotting and analysis

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-2.0, 3.0 + 1e-9, 0.001)
```

### Cell 2:Step and ramp definitions
- Define unit step and ramp functions
- Implement $ùë¢(ùë°)$ and $ùëü(ùë°)=ùë°$ ‚ãÖ $ùë¢(ùë°)$
 with vectorized operations.

- We choose $ùë¢(0)=1$
 to match MATLAB's default Heaviside behavior used in the solution

```python
def heaviside(x):
    """Unit step function u(t) with u(0)=1."""
    return np.where(x >= 0, 1.0, 0.0)

def ramp(x):
    """Ramp function r(t) = t * u(t)."""
    return x * heaviside(x)
```

### Cell 3: Signal construction:
- Construct the signal from ramp components
-  Build $ùë•(ùë°)=ùëü(ùë°)‚àíùëü(ùë°‚àí1)‚àíùëü(ùë°‚àí2)$
 exactly as in the provided solution.
- Each subtraction ‚Äúturns off‚Äù the growth after the corresponding breakpoint, shaping the piecewise-linear waveform.



```python
x = ramp(t) - ramp(t - 1) - ramp(t - 2)
```

### Cell 4: Plotting
- Plot the signal
- Visualize the result and match the MATLAB plot range and limits.
- Use grid and axis limits to clearly show the piecewise behavior.

```python
plt.figure(figsize=(7, 4))
plt.plot(t, x, color='navy', linewidth=2)
plt.title(r"$x(t) = r(t) - r(t-1) - r(t-2)$")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.ylim(-0.1, 1.1)
plt.xlim(-2, 3)
plt.grid(True, alpha=0.3)
plt.show()
```
