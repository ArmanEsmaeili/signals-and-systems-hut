      <div style="color: #7c3aed; font-weight: 700; margin-bottom: 8px;">SymPy</div>
      <div style="color: #64748b; font-size: 13px;">Symbolic mathematics and analytical computation</div>

    <div style="background: white; padding: 15px; border-radius: 8px; box-shadow: 0 2px 8px rgba(0,0,0,0.08);">
      <div style="color: #2563eb; font-weight: 700; margin-bottom: 8px;">NumPy</div>

```python
import sympy as sym
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Image, display

# ÿ™ŸÜÿ∏€åŸÖÿßÿ™ matplotlib
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['font.size'] = 11
plt.rcParams['axes.grid'] = True

print("‚úì Libraries imported successfully")
print("  - SymPy: Symbolic computation")
print("  - NumPy: Numerical operations")
print("  - Matplotlib: Visualization")
```

<div style="background: linear-gradient(135deg, #fef3c7 0%, #fde68a 100%); border-radius: 12px; padding: 25px; margin: 20px 0; border: 2px solid #fbbf24;">
  <h4 style="color: #92400e; margin-top: 0; display: flex; align-items: center; gap: 10px;">
    <span style="font-size: 24px;">üî£</span> Symbolic Signal Definition
  </h4>
  <p style="color: #78350f; line-height: 1.8;">
    The signal is defined symbolically using SymPy's <code style="background: #fef3c7; padding: 2px 8px; border-radius: 4px;">Heaviside</code> function. This allows mathematical manipulation and exact analytical evaluation.
  </p>
  <div style="background: #fffbeb; padding: 15px; border-radius: 8px; margin-top: 15px; border-left: 4px solid #f59e0b;">
    <p style="color: #78350f; margin: 0; font-family: monospace; font-size: 15px;">
      <strong>Expression:</strong> x(œÑ) = Heaviside(œÑ + 1) √ó Heaviside(œÑ - 1)
    </p>
  </div>
</div>

```python
# ÿ™ÿπÿ±€åŸÅ ŸÖÿ™ÿ∫€åÿ± ÿ≥ŸÖÿ®ŸàŸÑ€å⁄©
t, tau = sym.symbols('t tau', real=True)

# ÿ™ÿπÿ±€åŸÅ ÿ≥€å⁄ØŸÜÿßŸÑ ÿ®Ÿá ÿµŸàÿ±ÿ™ ÿ≥ŸÖÿ®ŸàŸÑ€å⁄©
# x(t) = u(t+1) * u(t-1)
xtau = sym.Heaviside(tau + 1) * sym.Heaviside(tau - 1)

print("="*60)
print("Signal Definition (Symbolic):")
print("="*60)
print("x(t) = u(t+1) ¬∑ u(t-1)")
print("\nSymPy Expression:")
sym.pprint(xtau)
print("="*60)
```

<div style="background: linear-gradient(135deg, #dbeafe 0%, #bfdbfe 100%); border-radius: 15px; padding: 30px; margin: 25px 0; border: 3px solid #3b82f6;">
  <h2 style="color: #1e40af; margin-top: 0; font-size: 26px; display: flex; align-items: center; gap: 12px;">
    <span style="background: #3b82f6; color: white; width: 40px; height: 40px; border-radius: 50%; display: flex; align-items: center; justify-content: center; font-weight: 700;">a</span>
    Part (a): Manual Signal Construction
  </h2>
  <p style="color: #1e3a8a; font-size: 16px; line-height: 1.8;">
    Construct the signal <strong>without</strong> using the <code style="background: #bfdbfe; padding: 3px 10px; border-radius: 5px;">heaviside</code> command by manually creating arrays of zeros and ones for different time regions.
  </p>
  
  <div style="background: white; padding: 20px; border-radius: 10px; margin-top: 20px; box-shadow: 0 4px 15px rgba(59, 130, 246, 0.2);">
    <h4 style="color: #1e40af; margin-top: 0;">Implementation Strategy:</h4>
    <ol style="color: #1e3a8a; line-height: 2;">
      <li>Divide time axis into three regions: before -1, between -1 and 1, after 1</li>
      <li>Create arrays: zeros for regions where x(t) = 0, ones where x(t) = 1</li>
      <li>Concatenate arrays to form complete signal</li>
    </ol>
  </div>
</div>

```python
# ============================================
# Part (a): ÿ®ÿØŸàŸÜ ÿßÿ≥ÿ™ŸÅÿßÿØŸá ÿßÿ≤ ÿØÿ≥ÿ™Ÿàÿ± heaviside
# ============================================
print("\n" + "="*60)
print("Part (a): Manual Construction (Without heaviside)")
print("="*60)

# ÿ®ÿßÿ≤Ÿá‚ÄåŸáÿß€å ÿ≤ŸÖÿßŸÜ€å
t1 = np.linspace(-5, -1, 100)
t2 = np.linspace(-1, 1, 100)
t3 = np.linspace(1, 10, 100)

# ÿ≥ÿßÿÆÿ™ ÿ≥€å⁄ØŸÜÿßŸÑ ÿ®Ÿá ÿµŸàÿ±ÿ™ ÿØÿ≥ÿ™€å
x1 = np.zeros(len(t1))  # ŸÇÿ®ŸÑ ÿßÿ≤ -1: x(t) = 0
x2 = np.zeros(len(t2))  # ÿ®€åŸÜ -1 Ÿà 1: x(t) = 0
x3 = np.ones(len(t3))   # ÿ®ÿπÿØ ÿßÿ≤ 1: x(t) = 1

# ÿ™ÿ±⁄©€åÿ® ÿ¢ÿ±ÿß€åŸá‚ÄåŸáÿß
t_manual = np.concatenate([t1, t2, t3])
x_manual = np.concatenate([x1, x2, x3])

print(f"‚úì Time vector created: {len(t_manual)} points")
print(f"  - Region 1 [-5, -1]: {len(t1)} points (zeros)")
print(f"  - Region 2 [-1,  1]: {len(t2)} points (zeros)")
print(f"  - Region 3 [ 1, 10]: {len(t3)} points (ones)")
print("="*60)

# ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ± Part (a)
plt.figure(figsize=(12, 5))
plt.plot(t_manual, x_manual, 'b-', linewidth=2.5, label=r'$x(t) = u(t+1) \cdot u(t-1)$')
plt.fill_between(t_manual, 0, x_manual, alpha=0.2, color='blue')

# ÿ™ŸÜÿ∏€åŸÖÿßÿ™ ŸÜŸÖŸàÿØÿßÿ±
plt.xlim((-5, 10))
plt.ylim((-0.1, 1.2))
plt.xlabel(r'Time $t$', fontsize=13, fontweight='bold')
plt.ylabel(r'$x(t)$', fontsize=13, fontweight='bold')
plt.title('Part (a): Signal Plot Without Heaviside Command',
          fontsize=15, fontweight='bold', pad=15)
plt.grid(True, alpha=0.3, linestyle='--')
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.8)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.8)
plt.axvline(x=-1, color='gray', linestyle=':', alpha=0.5, linewidth=1.5)
plt.axvline(x=1, color='gray', linestyle=':', alpha=0.5, linewidth=1.5)
plt.text(-1, 1.1, 't = -1', ha='center', fontsize=10, color='gray')
plt.text(1, 1.1, 't = 1', ha='center', fontsize=10, color='gray')
plt.legend(fontsize=11, loc='upper left')

plt.tight_layout()
plt.savefig("problem2_part_a.png", dpi=300, bbox_inches='tight')
plt.show()

print("‚úì Part (a) plot saved as 'problem2_part_a.png'")
```

<div style="background: linear-gradient(135deg, #fee2e2 0%, #fecaca 100%); border-radius: 15px; padding: 30px; margin: 25px 0; border: 3px solid #ef4444;">
  <h2 style="color: #991b1b; margin-top: 0; font-size: 26px; display: flex; align-items: center; gap: 12px;">
    <span style="background: #ef4444; color: white; width: 40px; height: 40px; border-radius: 50%; display: flex; align-items: center; justify-content: center; font-weight: 700;">b</span>
    Part (b): Using Heaviside Function
  </h2>
  <p style="color: #7f1d1d; font-size: 16px; line-height: 1.8;">
    Construct the signal using NumPy's built-in <code style="background: #fecaca; padding: 3px 10px; border-radius: 5px;">heaviside</code> function for cleaner, more mathematically intuitive implementation.
  </p>
  
  <div style="background: white; padding: 20px; border-radius: 10px; margin-top: 20px; box-shadow: 0 4px 15px rgba(239, 68, 68, 0.2);">
    <h4 style="color: #991b1b; margin-top: 0;">NumPy Heaviside Function:</h4>
    <p style="color: #7f1d1d; line-height: 1.8; margin-bottom: 10px;">
      <code style="background: #fee2e2; padding: 8px 12px; border-radius: 6px; display: block; font-size: 14px;">
        np.heaviside(x, value_at_zero)
      </code>
    </p>
    <ul style="color: #7f1d1d; line-height: 2;">
      <li><strong>First parameter:</strong> Input array</li>
      <li><strong>Second parameter:</strong> Value at x = 0 (discontinuity point)</li>
    </ul>
  </div>
</div>

```python
# ============================================
# Part (b): ÿ®ÿß ÿßÿ≥ÿ™ŸÅÿßÿØŸá ÿßÿ≤ ÿØÿ≥ÿ™Ÿàÿ± heaviside
# ============================================
print("\n" + "="*60)
print("Part (b): Using Heaviside Function")
print("="*60)

# ÿ®ÿßÿ≤Ÿá ÿ≤ŸÖÿßŸÜ€å ÿ®ÿß ÿØŸÇÿ™ ÿ®ÿßŸÑÿßÿ™ÿ±
t_heaviside = np.linspace(-5, 10, 500)

# ŸÖÿ≠ÿßÿ≥ÿ®Ÿá ÿ≥€å⁄ØŸÜÿßŸÑ ÿ®ÿß heaviside
# u(t+1) * u(t-1)
x_heaviside = np.heaviside(t_heaviside + 1, 1) * np.heaviside(t_heaviside - 1, 1)

print(f"‚úì Signal computed using NumPy heaviside function")
print(f"  - Time points: {len(t_heaviside)}")
print(f"  - Value at t=-1: {x_heaviside[np.argmin(np.abs(t_heaviside + 1))]}")
print(f"  - Value at t=1:  {x_heaviside[np.argmin(np.abs(t_heaviside - 1))]}")
print("="*60)

# ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ± Part (b)
plt.figure(figsize=(12, 5))
plt.plot(t_heaviside, x_heaviside, 'r-', linewidth=2.5, label=r'$x(t) = u(t+1) \cdot u(t-1)$')
plt.fill_between(t_heaviside, 0, x_heaviside, alpha=0.2, color='red')

# ÿ™ŸÜÿ∏€åŸÖÿßÿ™ ŸÜŸÖŸàÿØÿßÿ±
plt.xlim((-5, 10))
plt.ylim((-0.1, 1.2))
plt.xlabel(r'Time $t$', fontsize=13, fontweight='bold')
plt.ylabel(r'$x(t)$', fontsize=13, fontweight='bold')
plt.title('Part (b): Signal Plot Using Heaviside Command',
          fontsize=15, fontweight='bold', pad=15)
plt.grid(True, alpha=0.3, linestyle='--')
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.8)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.8)
plt.axvline(x=-1, color='gray', linestyle=':', alpha=0.5, linewidth=1.5)
plt.axvline(x=1, color='gray', linestyle=':', alpha=0.5, linewidth=1.5)
plt.text(-1, 1.1, 't = -1', ha='center', fontsize=10, color='gray')
plt.text(1, 1.1, 't = 1', ha='center', fontsize=10, color='gray')
plt.legend(fontsize=11, loc='upper left')

plt.tight_layout()
plt.savefig("problem2_part_b.png", dpi=300, bbox_inches='tight')
plt.show()

print("‚úì Part (b) plot saved as 'problem2_part_b.png'")
```
