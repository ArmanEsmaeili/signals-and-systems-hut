# === Cell 3 - Assumptions & method ===

### Assumptions
- **Signal:** \( x(t) = t \,\sin(2\pi t)\,[u(t) - u(t-3)] \).
- **Support:** Active only on \(0 \le t < 3\); zero elsewhere due to the unit steps.
- **Continuity:** The function is continuous on \([0,3)\) and equals zero outside this interval.
- **Sampling:** Use a uniform grid with small step \(dt\) for smooth plotting.

### Methods
- **Vectorized evaluation:** Compute \(t \sin(2\pi t)\) over the grid and multiply by the window mask.
- **Visualization:** Plot \(x(t)\) over a window that includes the entire support; mark the support boundaries at \(t=0\) and \(t=3\).


# === Cell 4 - Solution Section ===

**Imports:** Load NumPy for vectorized computation and Matplotlib for plotting.


```python
import numpy as np
import matplotlib.pyplot as plt
#Imports: Load NumPy for vectorized computation and Matplotlib for plotting.
```

```python
dt = 0.001
t_min, t_max = -1.0, 4.0
t = np.arange(t_min, t_max + dt, dt)    #Sampling and grid: Choose a fine step `dt = 0.001` and a window covering the support \([0,3]\) with margins.

```

```python
def u(z):
    return (z >= 0).astype(float)
#Unit step: Implement \(u(z)\) using a boolean mask converted to float.
```

```python
window = u(t) - u(t - 3.0)    #**Support window:** Build the time-limiting mask \([0,3)\) with \(u(t)-u(t-3)\).
```

```python
x = t * np.sin(2*np.pi*t) * window
#Signal definition:Compute \(x(t) = t \sin(2\pi t)\) and apply the window to zero-out values outside \([0,3)\).
```

```python
import matplotlib.pyplot as plt
#Plot: Visualize \(x(t)\) over the chosen window and mark the support boundaries at \(t=0\) and \(t=3\).
plt.figure(figsize=(8,3.5))
plt.plot(t, x, label=r'$x(t)=t\sin(2\pi t)\,[u(t)-u(t-3)]$')
plt.axvline(0, color='k', ls='--', alpha=0.35)
plt.axvline(3, color='k', ls='--', alpha=0.35)
plt.title('Time-limited sinusoidal ramp')
plt.xlabel('t'); plt.ylabel('x(t)')
plt.grid(alpha=0.3); plt.legend()
plt.show()
```
